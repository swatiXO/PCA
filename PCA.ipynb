# Importing libraries and dataset

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder, OneHotEncoder
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.decomposition import PCA
from sklearn.metrics import accuracy_score
from scipy.stats import zscore
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
df = pd.read_csv("heart.csv")
```

```python
df
```

# Removing outliers

```python
numerical_cols = df.select_dtypes(include=['number']).columns
```

```python
numerical_cols
```

```python
zScores = zscore(df[numerical_cols])
```

```python
zScores
```

```python
noOutliers = df[(zScores < 3).all(axis=1) & (zScores > -3).all(axis=1)]
```

```python
noOutliers
```

# converting text to numbers.

#Label Encoding for Binary Attributes

```python

binary_cols= []
for col in df.columns:
    if df[col].nunique() == 2:
        binary_cols.append(col)

print("Binary Columns:", binary_cols)
```

```python
labelEncoder = LabelEncoder()
```

```python
for col in binary_cols:
  noOutliers[col]= labelEncoder.fit_transform(noOutliers[col])
```

```python
noOutliers
```

#Using one hot encoding for categorical attributes.

```python
categorical_cols = noOutliers.select_dtypes(include=['object']).columns
categorical_cols
```

```python
noOutliers = pd.get_dummies(noOutliers, columns = categorical_cols,)
```

```python
noOutliers
```

# Scaling

```python
X= noOutliers.drop('HeartDisease', axis=1)
y = noOutliers['HeartDisease']
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
X_scaled
```

# Building our models

```python
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

```python
svm_model = SVC(gamma='scale',kernel = 'linear',C=1)
svm_model.fit(X_train, y_train)
y_pred = svm_model.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
conf_matrix = confusion_matrix(y_test, y_pred)
classification_rep = classification_report(y_test, y_pred)


print("Accuracy:", accuracy)
print("\nConfusion Matrix:\n", conf_matrix)
print("\nClassification Report:\n", classification_rep)
```

```python
logistic_model = LogisticRegression()
logistic_model.fit(X_train, y_train)

y_pred = logistic_model.predict(X_test)


accuracy = accuracy_score(y_test, y_pred)
conf_matrix = confusion_matrix(y_test, y_pred)
classification_rep = classification_report(y_test, y_pred)


print("Accuracy:", accuracy)
print("\nConfusion Matrix:\n", conf_matrix)
print("\nClassification Report:\n", classification_rep)
```

```python

rf_classifier = RandomForestClassifier(max_depth= None, min_samples_leaf= 1, min_samples_split= 10, n_estimators = 200)
rf_classifier.fit(X_train, y_train)
y_pred = rf_classifier.predict(X_test)


accuracy = accuracy_score(y_test, y_pred)
conf_matrix = confusion_matrix(y_test, y_pred)
classification_rep = classification_report(y_test, y_pred)


print("Accuracy:", accuracy)
print("\nConfusion Matrix:\n", conf_matrix)
print("\nClassification Report:\n", classification_rep)
```

# Random Forest giving me the best accuracy


# Now, applying PCA

```python

pca = PCA(n_components=len(X_scaled[0])//2)

X_pca = pca.fit_transform(X_scaled)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_pca, y, test_size=0.2, random_state=42)
```

```python
X_pca
```

```python
rf_classifier = RandomForestClassifier(max_depth= None, min_samples_leaf= 1, min_samples_split= 10, n_estimators = 200)
rf_classifier.fit(X_train, y_train)
y_pred = rf_classifier.predict(X_test)


accuracy = accuracy_score(y_test, y_pred)
conf_matrix = confusion_matrix(y_test, y_pred)
classification_rep = classification_report(y_test, y_pred)


print("Accuracy:", accuracy)
print("\nConfusion Matrix:\n", conf_matrix)
print("\nClassification Report:\n", classification_rep)
```

```python
# Display the explained variance ratio after PCA
print("\nExplained Variance Ratio after PCA:")
print(pca.explained_variance_ratio_)
```

```python
import numpy as np
cumulative_variance = np.cumsum(pca.explained_variance_ratio_)
print("Cumulative Explained Variance:", cumulative_variance)
```

# after PCA, approximately 78.8% of the original information is retained by our data.
